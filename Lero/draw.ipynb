```python
import matplotlib.pyplot as plt

# 定义函数来读取loss文件
def read_loss_file(file_path):
    epochs = []
    losses = []
    with open(file_path, 'r') as file:
        for line in file:
            if "Epoch" in line:  # 查找包含 Epoch 关键字的行
                parts = line.split(",")  # 分割行内容
                epoch_part = parts[0].strip()  # 获取 Epoch 部分
                loss_part = parts[1].strip()  # 获取 Loss 部分
                
                # 提取出数字
                epoch = int(epoch_part.split(":")[1])  # 例如: Epoch: 1 -> 1
                loss = float(loss_part.split(":")[1])  # 例如: training loss: 0.5 -> 0.5
                
                # 存储结果
                epochs.append(epoch)
                losses.append(loss)
    return epochs, losses
```

```python
# 读取两个loss文件
epochs1, losses1 = read_loss_file('../model/exp1/kmeans/16_30/target_loss.txt')
epochs2, losses2 = read_loss_file('../model/exp1/kmeans/16_30/random_0_loss.txt')
epochs3, losses3 = read_loss_file('../model/exp1/kmeans/16_30/random_1_loss.txt')
epochs4, losses4 = read_loss_file('../model/exp1/kmeans/16_30/random_2_loss.txt')
epochs5, losses5 = read_loss_file('../model/exp1/kmeans/16_30/random_3_loss.txt')
epochs6, losses6 = read_loss_file('../model/exp1/kmeans/16_30/random_4_loss.txt')


# 绘制曲线
plt.plot(epochs1, losses1, label='Target Model Loss', color='blue')
plt.plot(epochs2, losses2, label='random_0 Model Loss', color='orange')
plt.plot(epochs3, losses3, label='random_1 Model Loss', color='orange')
plt.plot(epochs4, losses4, label='random_2 Model Loss', color='orange')
plt.plot(epochs5, losses5, label='random_3 Model Loss', color='orange')
plt.plot(epochs6, losses6, label='random_4 Model Loss', color='orange')


# 图形标签
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Comparison of Training Loss for Two Models')
plt.legend()

# 显示图像
plt.show()
```
